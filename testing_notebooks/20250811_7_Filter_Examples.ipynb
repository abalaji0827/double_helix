```python
figsize = (1.5, 1.5)  # inches
dpi = 600
fig, ax = plt.subplots(figsize=figsize, dpi=dpi)

vmin_G2 = -0.005
vmax_G2 = 0.005

im_to_plot = new.g2a.T
vlag_cmap = sns.color_palette("vlag", as_cmap=True)
im = ax.imshow(im_to_plot, interpolation='none', cmap=vlag_cmap, vmin = vmin_G2, vmax=vmax_G2) 
ax.set_yticks([])
ax.set_xticks([])
ax.axis('off')
cbar = fig.colorbar(im, ax=ax, shrink=0.8, ticks=[vmin_G2, vmax_G2])
cbar.ax.tick_params(size=0)


scalebar_nm = 1000
scalebar_pix = scalebar_nm / 120
scalebar = AnchoredSizeBar(ax.transData,
                           scalebar_pix, '%d \u03BCm' % (scalebar_nm / 1e3), 'upper left', 
                           pad=0.05,
                           color='k',
                           frameon=False,
                           size_vertical=0.25)#,
                        #    fontproperties=fontprops)

ax.add_artist(scalebar)
plt.savefig(path.join(output_dir, "G2a.svg"), dpi=dpi, bbox_inches='tight')
```
